```python
import graphlab as gl
```

```python
sales=gl.SFrame('house_data.csv')
```

```python
sales
```

```python
train_d,test_d=sales.random_split(.8,seed=0)
```

```python
sqft_model=gl.linear_regression.create(train_d,target='price',features=['zipcode'])
```

```python
print(test_d['price'].mean())
```

```python
sqft_model.evaluate(test_d)
feat=['price','zipcode']
sales[feat].show()
```
